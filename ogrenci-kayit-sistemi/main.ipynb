```python
def basari_notu_hesapla():
    # Bilgileri alıyoruz
    ad = input("Adınızı giriniz        : ").upper()
    soyad = input("Soyadınızı giriniz     : ").upper()
    vize = float(input("Vize notunu giriniz    : "))
    final = float(input("Final notunu giriniz   : "))

    # Başarı notu hesaplama
    basari = vize * 0.3 + final * 0.7

    # Girdive çıktıları ekrana yazdırıyoruz
    print("Adınızı giriniz        : " + ad)
    print("Soyadınızı giriniz     : " + soyad)
    print("Vize notunu giriniz    : " + str(vize))
    print("Final notunu giriniz   : " + str(final))

    # Başarı notunu yazdırıyoruz
    print("Başarı notunuz         : ('" + ad + "', '" + soyad + "', " + str(basari) + ")")


# Fonksiyonu çalıştır
basari_notu_hesapla()
```

```python
def sayilari_incele(liste):
    tek_sayilar = []
    cift_sayilar = []

    # Sayıları tek ve çift olarak ayır
    for sayi in liste:
        if sayi % 2 == 0:
            cift_sayilar.append(sayi)
        else:
            tek_sayilar.append(sayi)

    # Tek sayılar için hesaplama
    tek_toplam = sum(tek_sayilar)
    tek_adet = len(tek_sayilar)
    if tek_adet > 0:
        tek_ortalama = tek_toplam / tek_adet
    else:
        tek_ortalama = 0

    # Çift sayılar için hesaplama
    cift_toplam = sum(cift_sayilar)
    cift_adet = len(cift_sayilar)
    if cift_adet > 0:
        cift_ortalama = cift_toplam / cift_adet
    else:
        cift_ortalama = 0

    # Tek sayıları ekrana yazdırıyoruz
    print("Tek sayılar:")
    print("Tek sayıların toplamı   :", tek_toplam)
    print("Tek sayı sayısı        :", tek_adet)
    print("Tek sayıların ortalaması:", tek_ortalama)

    # Çift sayıları yazdırıyoruz
    print("\nÇift sayılar:")
    print("Çift sayıların toplamı   :", cift_toplam)
    print("Çift sayı sayısı        :", cift_adet)
    print("Çift sayıların ortalaması:", cift_ortalama)

# Fonksiyonu çağırmak için listeyi veriyoruz
sayilar = [1, 22, 35, 77, 1305, 56, 98, 121, 1453, 640, 230, 61]
sayilari_incele(sayilar)
```

```python
class OgrenciKayitSistemi:
    def __init__(self):
        # öğrenci listesi tanımlanıyor
        self.ogrenci_listesi = []
        self.menu()  # Menü gösteriliyor

    def ogrenci_ekle(self):
        #  ad ve soyad alınıyor
        ad = input("Öğrencinin adını girin: ").strip().title()
        soyad = input("Öğrencinin soyadını girin: ").strip().title()

        if ad != "" and soyad != "":
            # Öğrenciyi ad ve soyadı birleştirerek listeye ekliyoruz
            ogrenci = ad + " " + soyad
            self.ogrenci_listesi.append(ogrenci)
            print(ogrenci + " başarıyla eklendi.\n")
        else:
            print("Ad ve soyad boş olamaz!\n")

    def ogrencileri_listele(self):
        # Liste boşsa uyarı ver
        if len(self.ogrenci_listesi) == 0:
            print("Henüz kayıtlı öğrenci yok.\n")
        else:
            print("\nKayıtlı Öğrenciler:")
            for i in range(len(self.ogrenci_listesi)):
                print(str(i + 1) + ". " + self.ogrenci_listesi[i])
            print("")

    def ogrenci_sil(self):
        # Kullanıcıdan silmek istediği öğrencinin adını ve soyadını alıyoruz
        ad = input("Silmek istediğiniz öğrencinin adını girin: ").strip().title()
        soyad = input("Silmek istediğiniz öğrencinin soyadını girin: ").strip().title()

        ogrenci = ad + " " + soyad

        if ogrenci in self.ogrenci_listesi:
            self.ogrenci_listesi.remove(ogrenci)
            print(ogrenci + " listeden silindi.\n")
        else:
            print("Öğrenci bulunamadı.\n")

    def menu(self):
        # Ana menü 
        while True:
            print("Öğrenci Kayıt Sistemi")
            print("[1] Öğrenci ekle")
            print("[2] Öğrencileri listele")
            print("[3] Öğrenciyi sil")
            print("[4] Programdan çık\n")

            secim = input("Seçiminizi giriniz (1/2/3/4): ").strip()
            print("")

            if secim == "1":
                self.ogrenci_ekle()
            elif secim == "2":
                self.ogrencileri_listele()
            elif secim == "3":
                self.ogrenci_sil()
            elif secim == "4":
                print("Programdan çıkılıyor...")
                break
            else:
                print("Geçersiz seçim. Lütfen tekrar deneyin.\n")

# Program  başlat
if __name__ == "__main__":
    sistem = OgrenciKayitSistemi()  # Sınıftan bir nesne oluşturuluyor
```
